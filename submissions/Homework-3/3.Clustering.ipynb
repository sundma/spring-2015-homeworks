## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

For exampe, the result of a clustering with 10 clusters is the following:


![A clustering of the locations of the calls for k=10](clusters10.png)

Filter the dataset to get the latitude and longitude, transform it as required and run the k-means clustering algorithm. Argue about the number of clusters you used. **(pts 50)**

```python
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))


def numNeighborhoods():
    n = []
#print( boston['neighborhood'] )

    for  i in boston['neighborhood']:
        print i
        if i not in n:
            n.append(i)

    print len(n)
    
#Number of neighborhoods = 24  


locations = pd.concat([boston['LATITUDE'], boston['LONGITUDE']], axis=1, keys=['LATITUDE', 'LONGITUDE'])
scaled_locations = preprocessing.StandardScaler().fit_transform(locations)

cmap = sns.color_palette("dark", n_colors=10)
cmap.extend(sns.color_palette("bright", n_colors=14))

#sns.hls_palette(n_colors= 14 , h=0.7, l=0.4, s=0.8))

kmeans = KMeans(init = 'k-means++', n_clusters = 24, n_init = 10)
kmeans.fit_predict(scaled_locations)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

print "The total error of the clustering is: ", error
print '\nCluster labels'
print labels
print '\n Cluster Centroids'
print centroids


scaled_X = scaled_locations[:, 1]
scaled_Y = scaled_locations[:, 0]



for i, label in enumerate(set(labels)):
    point_indices = np.where(labels == label)[0]
    point_indices = point_indices.tolist()
    plt.scatter(scaled_X[point_indices], scaled_Y[point_indices], s=40, alpha=0.2, c=cmap[i], marker='+')
```

```python
evaluate_clusters(scaled_locations,40) 
```

I chose 24 clusters for two reasons. 
1. There are 24 unique neighborhoods in the data set, and neighborhoods are assigned based on geographic location, therefore it seems likely that the data would fall nicely into 24 clusters. 
2. As can be seen in the graph above the error stabalizes around 20 clusters. Although you could pick 20 clusters without a very large increase in error (2837.43279328 (20) vs 2293.18807026 (24) ); however, basec on visual inspection and the fact that there are 24 neighborhoods I chose 24 for the number of clusters. 

## Step 2. Using multiple features

The next step in our analysis is to use more features than just the geolocation of a record (whose clustering only gives us information about the neighborhoods, which should not be a huge revelation!) Filter the dataset to get more features, transform them as required and compute another clustering.
```python
boston = pd.read_csv('../Homework-2/cases_2015.csv')

fig, ax = plt.subplots(1,1,figsize=(10,10))


#Clustering location and on time status 

#Replace string values with binary representation 
boston.OnTime_Status[boston.OnTime_Status == 'ONTIME' ] = 0.0
boston.OnTime_Status[boston.OnTime_Status == 'OVERDUE' ] = 1.0

#Make the dataframe using location from part 1 
features = pd.concat([locations, boston['OnTime_Status']], axis=1)
features_filtered = features[features.OnTime_Status.notnull()]

#Normalize the data 
scaled_features = preprocessing.StandardScaler().fit_transform(features_filtered)

#Colors for graphing 
cmap = sns.color_palette("dark", n_colors=11)
cmap.extend(sns.color_palette("bright", n_colors=14))

#Run K means with 24 clusters
kmeans = KMeans(init = 'k-means++', n_clusters = 25, n_init = 10)
kmeans.fit_predict(scaled_features)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

print "The total error of the clustering is: ", error
print '\nCluster labels'
print labels


scaled_X = scaled_features[:, 1]
scaled_Y = scaled_features[:, 0]



for i, label in enumerate(set(labels)):
    point_indices = np.where(labels == label)[0]
    point_indices = point_indices.tolist()
    plt.scatter(scaled_X[point_indices], scaled_Y[point_indices], s=40, alpha=0.2, c=cmap[i], marker='+')
```

This cluster represents the relationship between geographic location an whether the case is ontime or overdue. I thought that there might be a relationship between these two things based on inner city or suburb or high priority / low priority area. The error for this clustering is low compared to some other clusterings I tried (see below), so I feel that there is at least somewhat of a relationship between these two things. The areas closer to the center of the city are clustered more variably, but the outer areas appear to clustered more in terms of area, which could be due to the fact that locations further from the city are resolved after inner city cases, or that in general cases are resolved by area, and cases are generated more quickly where it is more populated (inner city), resulting in the variability within the city but better clustering in the suburbs. 

```python
evaluate_clusters(scaled_features, 40) 
```

## Step 3. Detecting outliers

Now that we have a clustering of the records in the dataset, we can detect the outlying ones. These are the ones who are the farthest from the centroids of their clusters. Track them down and describe any interesting observations that you can make. **(pts 20)** 

```python
#Using clustering of location and on time status 

#Run K means with 24 clusters
kmeans = KMeans(init = 'k-means++', n_clusters = 24, n_init = 10)
kmeans.fit_predict(scaled_features)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

for index in range(0, 24):
    #For each cluster get the indices of its members in the data set 
    indices = [i for i, x in enumerate(labels) if x == index]
    
    #get the centroid 
    centroid = centroids[index]
    
    #init 
    largestOutlier = metrics.euclidean_distances(centroid, scaled_features[0])[0][0]
    largestOutlierIndex = 0
    
    #go through each entry for this cluster 
    for i in indices: 
         d = metrics.euclidean_distances(centroid, scaled_features[i])
            
         #update largest distance if needed  
         if d[0][0] > largestOutlier:
                largestOutlier = d[0]
                largestOutlierIndex = i
                
                
    print "\nCentroid",index,": ", centroid
    
    print "Largest Outlier: ", scaled_features[largestOutlierIndex],", ",largestOutlier
    
    
    
                                    
                                   
        


```

It seems that in some cases the largest outliers were caused by a difference in location (being physically far from the other cases), and in other cases it was caused by having the opposite case status (ontime vs overdue). However, outliers were caused by being geographically far from the centroid in more cases than outliers being overdue when they were supposed to ontime or vice versa. I found it interesting that there were no outliers (of the largest) that had a case status of ontime. This may have to do with the ratio of ontime to overday statuses in the dataset.
