# Part 2: Can we predict whether someone gives positive or negative reviews?

Now that we know that users who have only written one review tend to give a very positive review or a very negative review, let's see if we can accurately predict whether someone gives a positive or negative review based on the information we can find in their profile. In order to simplify the problem a bit, we are going to say a user gives 'positive' reviews if their average star count is between 3.5 (inclusive) and 5, and that a user gives 'negative' reviews if theur average star count is between 1 and 3.5 (exclusive). In order to proceed the data first needs some preprocessing. The average_stars field needs to be converted to a binary value. Additionally, it would probably be more meaningful to convert the 'yelping_since' to the number of days a person has been yelping instead of the date that they started. 

```python

def findDaysSince(d):
    (year, month) = d.split("-")
    d1 = date(2015,4,1)
    d2 = date(int(year), int(month), 1)
    
    return (d1-d2).days 
    



user['yelping_since'] = user['yelping_since'].apply(lambda x: findDaysSince(x) ) 
user =user.rename(columns = {'yelping_since':'days_yelping'})

```

Now let's drop some of the columns that we aren't interested in, like name, user_id, elite, type...

```python
user2 = user.drop(['name', 'type', 'elite', 'user_id', 'compliments.list', 'compliments.note', 'compliments.plain', 'compliments.profile', 'compliments.more', 'compliments.list', 'compliments.photos'  ], axis=1, inplace=False)
```

Next we'll replace all of the NaN values with 0. Convert the friends list column to a number of friends column, and convert the average starts column to a binary column, this will serve as the column we are trying to predict via logistical regression 

```python

def findNumFriends(x):
    if x  == '[]':
        return 0
    else: 
        
        return len(x.split(','))
    
def posNeg(x):
    if x >= 3.5:
        return 1
    else:
        return 0
 
user2['friends'] = user2['friends'].apply(lambda x: findNumFriends(x))
user2['average_stars'] = user2['average_stars'].apply(lambda x: posNeg(int(x)))
user2 = user2.fillna(0)
```

The data is almost ready. Upon examining the data it appears that there isn't much data for users who have written less than 40 reviews. Therefore we will remove the users who haven't written at least 40 reviews, as we just don't have enough data about them. 

```python
userData = user2[user2.review_count > 40]
```

Logistic Regression...

```python
labels = list(userData.columns.values)
l1  = labels[:8]
l2 = labels[9:]
labels = l1+l2


y = userData['average_stars']
X = userData.drop(['average_stars'], axis = 1)

X = preprocessing.StandardScaler().fit_transform(X)

X, y = utils.shuffle(X, y, random_state=1)
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.3, random_state=0)

logistic_regr = linear_model.LogisticRegression()
logistic_regr.fit(X_train, y_train)
y_predicted = logistic_regr.predict(X_test)
                
probs = logistic_regr.predict_proba(X_test)
print "Score: ",metrics.accuracy_score(y_test, y_predicted)  

print pd.DataFrame(zip(labels, np.transpose(logistic_regr.coef_)))
```

#  Part 3: Location, Location, Location? 

Is there a relationship between the quality of a restaurant and its geographic location? 
For example, will we find that high quality restaurants are found in upscale or trendy areas, while low quality restaurants are found in less desirable areas?

```python

def restaurants(x):
    if 'Restaurants' in x:
        return 1
    else:
        return 0

businessWI = business[business['city'] == 'Madison']
businessWI['categories'] = businessWI['categories'].apply(lambda x:  restaurants(x) )

restaurantWI = businessWI[businessWI['categories'] == 1]
restaurantWI
WI = pd.concat([restaurantWI['longitude'], restaurantWI['latitude'], restaurantWI['stars']], axis=1, keys=['longitude', 'latitude', 'stars'])

X = WI['longitude']
X = X.as_matrix()
Y = WI['latitude']
Y = Y.as_matrix()
```

```python
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')
```

```python
scaled_features = preprocessing.StandardScaler().fit_transform(WI)

evaluate_clusters(scaled_features, 15)
```

```python
#clustering


fig, ax = plt.subplots(1,1,figsize=(10,10))

kmeans = KMeans(init = 'k-means++', n_clusters = 5, n_init = 10)
labels = kmeans.fit_predict(scaled_features)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

print "The total error of the clustering is: ", error

cmap = sns.color_palette("bright", n_colors=4)
cmap.extend(sns.hls_palette(n_colors=8, h=0.7, l=0.4, s=0.8))


scaled_X = scaled_features[:, 1]
scaled_Y = scaled_features[:, 0]



f = open('WI.csv', 'w')

f.write('longitude, latitude, stars, cluster')

# Iterate over the different clustering labels, find the points
    # that belong to that cluster and plot them with the same color
for i, label in enumerate(set(labels)):
    averageR = 0;
    point_indices = where(labels == label)[0]
    point_indices = point_indices.tolist()
    
    
    for index in point_indices:
        stars = restaurantWI.iloc[[index]].stars
        averageR += float(stars)
    a = averageR / len( point_indices)    
    for point in point_indices:
        
        row = str(X[point])+","+str(Y[point])+","+str(a)+","+str(label)+"\n"
        f.write(row)
        
        
    print label,": ", (a)
        
    ax.scatter(X[point_indices], Y[point_indices], s=35, 
                    alpha=0.4, c=cmap[i])
    
```

# Part 4: How to Stay in Business 

About 20% of the restaurants in this dataset are now closed. If you're opening a business or want to stay in business, what things should you do, or not do, to increase your chance of success? We'll use logistic regression again to see if we can predict whether a business is open or closed, and try to find some interesting attributes that contribute.
