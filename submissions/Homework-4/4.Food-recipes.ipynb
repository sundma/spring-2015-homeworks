### Chemical Compounds in Recipes

Do recipes from the same region (example: North America -> Canada, Cajun Creole, American, Southern_Soul Food, Southwestern) have more similar chemical compounds than recipes from other regions? It may be true that recipes from similar geographic locations are more likely to incorporate the same chemical compounds, due to the use of similar ingredients; however, it may also be true that there are a set of common chemical compounds that are found in most recipes, regardless of the region of origin. In order to investigate this question the top ten most commonly occuring chemical compounds will be determined for each country/cuisine, and then this data will be clustered in order to determine whether there is a relationship between location and chemical compounds found in recipes. It is expected that cuisines that utilize many of the same compounds will be clustered together sooner.

### Hypothesis 

Clusters based on  geographic location (NorthAmerican, MiddleEastern, EastAsian, LatinAmerican, SouthAsian, African, SouthernEuropean, EasternEuropean, SoutheastAsian, WesternEuropean, NorthernEuropean), will be visible in the clustered data.

### Clustering

```python

#Custom method to produce a distance matrix based on the number of compounds
#two entries have in common

#

m = c10.as_matrix() 
distanceMatrix = squareform(pdist(c10,'hamming'))

for i in range(0, len(m)):
    c1 = m[i]
    for j in range(0, len(m)):
        if i == j: 
            distanceMatrix[i][j] = 0.0
        else:
            c2 = m[j]
            sim = 0.0
            for h in range(0, len(m[i])):
                for k in range(0, len(m[i])):
                    
                    if c1[h] == c2[k]:
                        sim += 0.1
            distanceMatrix[i][j] = (1-sim)
      
      
  
```

```python
#Clustering 
#using inline plt gives me an error but I can see the plot. The inline command can be commented out
#to avoid the error. 


Z = hr.linkage(distanceMatrix, method='complete')
fig = plt.figure(figsize=(10,10))
T = hr.dendrogram(Z, color_threshold=1.1, leaf_font_size=5)
fig.show()
```
