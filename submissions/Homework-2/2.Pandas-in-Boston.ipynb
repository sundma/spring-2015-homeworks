Make a plot (similar to the one above, like a horizontal histogram) that for each neighborhood shows the number of snow-related calls made from there. **(5 pts)**

````python
neighborhoods = {}

snow =  boston[boston['CASE_TITLE'].str.contains("Snow", na = False)]

for i, row in snow.iterrows():
    n = row.neighborhood 
    if n in neighborhoods:
        neighborhoods[n] +=1
    elif str(n) != 'nan':
        neighborhoods[n] = 1 

plt.barh(range(len(neighborhoods)), neighborhoods.values())
plt.yticks(range(len(neighborhoods)), neighborhoods.keys())
plt.show()





    
````

## Step 3. Advanced filtering & joins

By now, you should have noticed that, for each day, there are multiple records in our dataframe. Use the `groupby` function to get, for each day, the counts for all the case categories. Plot the result for the top 5 categories as you see fit. **(20 pts)**

````python
# Write your code here


caseCounts =  boston.groupby("CASE_TITLE").size().order()
caseNames = caseCounts.index
top5Titles = [caseNames[-1], caseNames[-2], caseNames[-3], caseNames[-4], caseNames[-5]]

# lists to hold the number of cases for the top five per day 
first = []
second = []
third = []
fourth = []
fifth = []

days = []

i = 0

#boston.OPEN_DT = boston.OPEN_DT.apply(lambda d: d.date() )

top5 = boston[boston['CASE_TITLE'].isin(top5Titles)]

for day, dayData in top5.groupby("OPEN_DT"):

    
    
    days.append(str(day))
    first.append(0)
    second.append(0)
    third.append(0)
    fourth.append(0)
    fifth.append(0)
    
    for idx, row in dayData.iterrows():
    
        if str(row.CASE_TITLE) == top5Titles[0]:
            first[i] += 1
        elif str(row.CASE_TITLE) == top5Titles[1]:
            second[i] += 1
        elif str(row.CASE_TITLE) == top5Titles[2]:
            third[i] += 1
        elif str(row.CASE_TITLE) == top5Titles[3]:
            fourth[i] += 1
        elif str(row.CASE_TITLE) == top5Titles[4]:
            fifth[i] += 1
    i += 1

top5df = pd.DataFrame({"Date" : days, top5Titles[0]: first, top5Titles[1]: second, top5Titles[2]: third, top5Titles[3]: fourth, top5Titles[4]:fifth})

top5df.plot(x = 'Date', figsize=(12,5) )    


````

Let's see how this trend compares to the snowfall. First, we need to read the data from `snow.csv`

````python
# Read the dataframe
snow_data = pd.read_csv("snow.csv",parse_dates=[1])
````

According to the documentation, the column `SNOW` has the values for the snowfall, in mm. However, the dataset we have has information from many different stations around Boston. Aggregate all their measurements and transform the result to centimeters. Finally, plot the result **(10 pts)**

````python
# Write your code here
#drop the rows with negative snow fall because it is not clear if they are intended to be 0



snowClean = snow_data[snow_data.SNOW >= 0] 
snowFall = {}

for day, dayData in snowClean.groupby("DATE"):
    snowFall[str(day.date())] = (dayData.SNOW.mean()) * 0.1

   
avgSnow = pd.DataFrame(snowFall.items(), columns=['Date', 'Average Snow (cm)'])

snow = avgSnow.sort('Date')
snow.reset_index(drop = True)

snow.plot(x = 'Date', figsize=(12,5) )


    
        
````

## Step 4. Matching trends

The last thing we will do is to join the snowfall dataframe with that of the top 5 categories of calls, per day. Don't forget to fill the NaN values with 0. You want to do an outer join (if there are calls for a day but no snowfall data, we want to create a 0 record). Save that to a dataframe called `calls_on_snow`. The first 5 columns should be the counts for the categories, and the 6th will be the snowfall data. **(10 pts)**

````python


calls_on_snow = pd.merge(top5df, snow, on = 'Date')

calls_on_snow.ix[0,0]
````

Now, you can show the connection of the two trends (the calls and the snow) in a plot. I provide some sample code, but you should also try your own versions.

````python
fig, axes = plt.subplots(nrows=2,figsize=(12,5), sharex=True)

calls_on_snow.ix[:,1:6].plot(ax=axes[0])
calls_on_snow.ix[:,6].plot(ax=axes[1],label="snowfall (in cm)", color='gray')

axes[0].set_title("The effect of the winter snowstorms on Boston's 311 Hotline for 2015")
axes[0].set_ylabel("# Calls")
axes[1].set_ylabel("Snowfall in cm")
axes[1].set_xlabel("Date")
axes[0].xaxis.grid(False)
axes[1].xaxis.grid(False)
````
